## 2. The question: decision, action, cost of a wrong call

**The question:** Which content pages should a review team look at first, given limited time to act on refresh and optimization opportunities?

**The decision this improves:** Right now, without a data-driven ranking, a content team might review pages by gut feeling, by age, or randomly. This project gives them a ranked list backed by evidence instead — so limited review time goes to the pages most likely to matter.

**Who acts on it:** A content strategist or SEO reviewer, who picks the top pages from the ranked queue and decides whether to refresh, expand, monitor, or leave each one alone, based on the reason codes attached to it.

**Cost of a wrong recommendation:**
- **False positive** (flagged as worth reviewing, but wasn't actually a good candidate): the team wastes limited review time on a low-value page, while a genuinely declining, high-traffic page might get missed instead.
- **False negative** (a real decline goes unflagged): that page keeps losing visibility and traffic unnoticed, until it's much harder — or too late — to recover.

**Why data or ML helps at all:** With thousands of content pages, a human can't manually track every trend, position shift, or engagement change across the whole inventory. A model can process signals across all pages at once and surface the ones most worth a human's limited attention — while the human still makes the final call on each one.
